```python
import os
```

```python
list_town = os.listdir('/mnt/data/PETA_dataset')
list_town.sort()
list_towns = [f'/mnt/data/PETA_dataset/{town}/archive' for town in list_town]
list_labels = [town.replace('archive','Label.txt') for town in list_towns]
list_process_towns = [f'/mnt/data/PETA_dataset/{town}/processed' for town in list_town]
print(list_process_towns)
for town in list_process_towns:
    if not os.path.exists(town):
        os.mkdir(town)
```

```python
import json
from PIL import Image
import random
for p_id in range(0,10):
    annotation = {}
    Collect_keys = ['Black', 'Grey', 'Blue', 'White', 'Brown', 'Red', 'Purple', 'Green', 'Pink', 'Orange', 'Yellow']
    
    print(list_labels[p_id])
    with open(list_labels[p_id], 'r') as f:
        labels = f.readlines()
        # print(labels[0])
        
        labels = [l.split(' ') for l in labels]
        for l in labels:
            image_id = l[0] 
            
            upper_color = next((i for i, x in enumerate(l) if 'upperBody' in x and any(keyword in x for keyword in Collect_keys)), -1)

            lower_color = next((i for i, x in enumerate(l) if 'lowerBody' in x and any(keyword in x for keyword in Collect_keys)), -1)
            annotation[image_id] = {
                "upper": None if upper_color == -1  else l[upper_color].replace('upperBody', ''),
                "lower": None if lower_color == -1  else l[lower_color].replace('lowerBody', ''),
            }
    js_anno = []
    list_images = os.listdir(list_towns[p_id])
    test_path = random.sample(list_images, int(len(list_images)*0.15))

    for i, image_name in enumerate(list_images):
        path = list_towns[p_id]+'/' + image_name
        image = Image.open(path)

        # Ly kch thc nh
        width, height = image.size
        # Ct nh lm i theo chiu ngang
        top_half = image.crop((0, 0, width, height // 2))
        bottom_half = image.crop((0, height // 2, width, height))
        # print(list_process_towns[p_id])
        top_half.save(f"{list_process_towns[p_id]}/top_{image_name}")
        bottom_half.save(f"{list_process_towns[p_id]}/bot_{image_name}")
        id = image_name.split('_')
        id = id[0]

        js_anno.append({
            'image': f"{list_process_towns[p_id]}/top_{image_name}",
            'color': annotation[id]['upper'],
            'is_train': False if image_name in test_path else True
        })
        js_anno.append({
            'image': f"{list_process_towns[p_id]}/bot_{image_name}",
            'color': annotation[id]['lower'],
            'is_train': False if image_name in test_path else True
        })
    print(js_anno)
    json_file = list_labels[p_id].replace('Label.txt','Label_process.json')
    print(list_labels[p_id].replace('Label.txt','Label_process.json'))
    with open(json_file, 'w') as f:
        json.dump(js_anno, f, indent=4)
```
